```python
import os
import base64
import mimetypes
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage


# =========================
# CONFIGURATION
# =========================
MODEL_NAME = "qwen3-vl:235b-instruct-cloud"
# For local testing use smaller model:
# MODEL_NAME = "qwen2.5-vl:7b"


# =========================
# UTIL FUNCTIONS
# =========================

def image_to_data_url(path: str) -> str:
    """
    Converts an image file to base64 data URL.
    Automatically detects MIME type.
    """
    if not os.path.isfile(path):
        raise FileNotFoundError(f"File not found: {path}")

    mime_type, _ = mimetypes.guess_type(path)
    if mime_type is None:
        mime_type = "image/jpeg"

    with open(path, "rb") as f:
        encoded = base64.b64encode(f.read()).decode("utf-8")
    return f"data:{mime_type};base64,{encoded}"


def build_multimodal_message(prompt: str, image_paths: list):
    """
    Builds a LangChain HumanMessage with multiple images.
    """
    content = [{"type": "text", "text": prompt}]

    for path in image_paths:
        data_url = image_to_data_url(path)
        content.append({
            "type": "image_url",
            "image_url": data_url
        })

    return HumanMessage(content=content)


# =========================
# MAIN FUNCTION
# =========================

def ask_vision_model(prompt: str, image_paths: list):
    model = ChatOllama(model=MODEL_NAME)

    message = build_multimodal_message(prompt, image_paths)

    print("\n🧠 Sending request to model...\n")
    response = model.stream([message])
    result=""
    for chunk in response:
        print(chunk.content, end="", flush=True)
        result += chunk.content
    return result

# =========================
# RUN EXAMPLE
# =========================

if __name__ == "__main__":
    prompt = "write a research paper on the basis of these images"

    images = [
       r"C:\Users\miyas\OneDrive\Pictures\Screenshots\Screenshot 2026-02-08 140337.png",
       r"C:\Users\miyas\OneDrive\Pictures\Screenshots\Screenshot 2026-02-09 121942.png",
    ]

    try:
        print("\n📸 MODEL RESPONSE:\n")
        x=ask_vision_model(prompt, images)
        print("\n\n✅ Final Result:\n")
        print(len(x))
    except Exception as e:
        print("\n❌ ERROR:", e)
```
