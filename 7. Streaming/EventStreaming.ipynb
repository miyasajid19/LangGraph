````python
from langgraph.graph import StateGraph,START,END
from langchain_ollama import ChatOllama
from typing import TypedDict
from rich import print
class State(TypedDict):
    message:str
    response:str

model = ChatOllama(model="minimax-m3:cloud", reasoning=True)

def chat_node(state:State):
    print("Chat node received state:", state)
    message = state['message']
    response=model.invoke(message)
    return {"response": response}


graph=StateGraph(State)
graph.add_node("chat", chat_node)
graph.add_edge(START, "chat")
graph.add_edge("chat", END)
app=graph.compile()


stream = app.stream_events({
    "message": "1+1"
}, version="v3")

for message in stream.messages:
    # lets say there are 4 nodes which will stream and message will stream message will be looped 4 times and we want to print the message in each node
        print("Reasoning:", end=" ", flush=True)
        for token in message.reasoning:
            print(token, end="", flush=True)    
        print("Tool calls")
        for token in message.tool_calls:
            print(token, end="", flush=True)
        print("\nMessage:", end=" ", flush=True)
        for token in message.text:
            # this will be for one particular node, if there are 4 nodes then this will be looped 4 times and we can print the token in each node
            print(token, end="", flush=True)
````

````python
from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama
from typing import TypedDict


class State(TypedDict):
    topic: str
    joke: str
    poem:str
    explaination: str


model = ChatOllama(
    model="minimax-m3:cloud",
    reasoning=True
)


def joke_node(state: State):
    response = model.invoke(
        f"Write a joke about {state['topic']}. Return only the joke."
    )
    return {"joke": response.content}

def poem_node(state: State):
    response = model.invoke(
        f"Write a poem about {state['topic']}. Return only the poem."
    )
    return {"poem": response.content}


def explaination_node(state: State):
    response = model.invoke(
        f"Explain the following joke in detail:\n\n{state['joke']}"
    )
    return {"explaination": response.content}


builder = StateGraph(State)

builder.add_node("joke_node", joke_node)
builder.add_node("poem_node", poem_node)
builder.add_node("explaination_node", explaination_node)

builder.add_edge(START, "joke_node")
builder.add_edge(START, "poem_node")
builder.add_edge("joke_node", "explaination_node")
builder.add_edge("poem_node", "explaination_node")
builder.add_edge("explaination_node", END)

graph = builder.compile()
async def main():

    async for event in graph.astream_events(
        {"topic": "Transformers"},
        version="v2",
    ):

        event_type = event["event"]

        if event_type == "on_chain_start":
            print("\nSTART:", event["name"])

        elif event_type == "on_chain_end":
            print("\nEND:", event["name"])

        elif event_type == "on_chat_model_stream":
            chunk = event["data"]["chunk"]

            if chunk.content:
                print(chunk.content, end="", flush=True)

await main()
````

````python
from typing import TypedDict, Literal

from pydantic import BaseModel
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from langgraph.graph import StateGraph, START, END

# ============================================================
# State
# ============================================================

class State(TypedDict, total=False):
    query: str

    joke: str
    poem: str

    summary_joke: str
    summary_poem: str
    summary: str

    suggestion: str


# ============================================================
# Router Schema
# ============================================================

class Schema(BaseModel):
    path: Literal[
        "joke_summary",
        "poem_summary",
        "both_summary",
    ]


parser = PydanticOutputParser(pydantic_object=Schema)


# ============================================================
# Model
# ============================================================

model = ChatOllama(
    model="gemma4:31b-cloud",
    reasoning=True,
    format="json",
)


# ============================================================
# Router Prompt
# ============================================================

router_prompt = PromptTemplate.from_template(
    """
You are a routing assistant.

The joke and poem already exist.

Decide which summary node should run:

- joke_summary -> summarize only the joke
- poem_summary -> summarize only the poem
- both_summary -> summarize both

User Query:
{query}

{format}
""",
    partial_variables={
        "format": parser.get_format_instructions()
    },
)


# ============================================================
# Nodes
# ============================================================

def joke_node(state: State):
    response = model.invoke(
        f"Write a joke about {state['query']}."
    )

    return {
        "joke": response.content
    }


def poem_node(state: State):
    response = model.invoke(
        f"Write a poem about {state['query']}."
    )

    return {
        "poem": response.content
    }


def router_node(state: State):
    # Dummy node used for synchronization
    return {}


def determine_path(state):
    response = model.invoke(
        router_prompt.format(query=state["query"])
    )

    content = response.content

    if isinstance(content, list):
        text = "".join(
            block["text"]
            for block in content
            if block.get("type") == "text"
        )
    else:
        text = content

    print(text)

    parsed = parser.parse(text)

    return parsed.path


def summarize_joke(state: State):
    response = model.invoke(
        f"Summarize the following joke:\n\n{state['joke']}"
    )

    return {
        "summary_joke": response.content
    }


def summarize_poem(state: State):
    response = model.invoke(
        f"Summarize the following poem:\n\n{state['poem']}"
    )

    return {
        "summary_poem": response.content
    }


def summarize(state: State):
    response = model.invoke(
        f"""
Summarize the following joke and poem.

Joke:
{state['joke']}

Poem:
{state['poem']}
"""
    )

    return {
        "summary": response.content
    }


def final_node(state: State):
    response = model.invoke(
        f"""
Suggest the next topic based on this state.

{state}
"""
    )

    return {
        "suggestion": response.content
    }


# ============================================================
# Build Graph
# ============================================================

builder = StateGraph(State)

builder.add_node("joke_node", joke_node)
builder.add_node("poem_node", poem_node)

builder.add_node("router", router_node)

builder.add_node("summarize_joke", summarize_joke)
builder.add_node("summarize_poem", summarize_poem)
builder.add_node("summarize", summarize)

builder.add_node("final_node", final_node)


# ============================================================
# Parallel generation
# ============================================================

builder.add_edge(START, "joke_node")
builder.add_edge(START, "poem_node")

# Wait for both branches to complete
builder.add_edge("joke_node", "router")
builder.add_edge("poem_node", "router")


# ============================================================
# Routing
# ============================================================

builder.add_conditional_edges(
    "router",
    determine_path,
    {
        "joke_summary": "summarize_joke",
        "poem_summary": "summarize_poem",
        "both_summary": "summarize",
    },
)


# ============================================================
# Final
# ============================================================

builder.add_edge("summarize_joke", "final_node")
builder.add_edge("summarize_poem", "final_node")
builder.add_edge("summarize", "final_node")

builder.add_edge("final_node", END)


# ============================================================
# Compile
# ============================================================

graph = builder.compile()
import asyncio

stream = await graph.astream_events(
    {
        "query": "think of shape of pizza and write joke and poem and summarize them and suggest next topic"
    },
    version="v3",
)


async def consume_messages():
    async for message in stream.messages:
        print(f"\n[llm] node={message.node}")
        # reasoning
        if hasattr(message, "reasoning"):
            print("\n\n[reasoning]")
            async for token in message.reasoning:
                print(token, end="", flush=True)

        # generated text
        if hasattr(message, "text"):
            async for token in message.text:
                print(token, end="", flush=True)


        # tool calls
        if hasattr(message, "tool_calls"):
            print("\n\n[tool_calls]")
            async for call in message.tool_calls:
                print(call)


async def consume_values():
    async for value in stream.values:
        print(
            f"\n[value] node={value}"
        )


async def consume_subgraphs():
    async for subgraph in stream.subgraphs:
        print(
            f"\n[subgraph] path={subgraph.path}"
        )


await asyncio.gather(
    consume_messages(),
    consume_values(),
    consume_subgraphs(),
)

print("\n\nFinal State:")
print(stream.output)
````

````python
from typing import TypedDict

from langchain_ollama import ChatOllama
from langgraph.graph import StateGraph, START, END


# ------------------------------------------------------------------
# State
# ------------------------------------------------------------------

class State(TypedDict, total=False):
    query: str

    joke: str
    poem: str

    summary: str
    suggestion: str


# ------------------------------------------------------------------
# Model
# ------------------------------------------------------------------

model = ChatOllama(
    model="gemma4:31b-cloud",
)


# ------------------------------------------------------------------
# Main Graph Nodes
# ------------------------------------------------------------------

def joke_node(state: State):
    response = model.invoke(
        f"Write a short joke about {state['query']}."
    )

    return {
        "joke": response.content
    }


def summarize(state: State):
    response = model.invoke(
        f"""
Summarize these results.

Joke:
{state.get("joke", "")}

Poem:
{state.get("poem", "")}
"""
    )

    return {
        "summary": response.content
    }


def final_node(state: State):
    response = model.invoke(
        f"""
Suggest a related topic.

Current state:

{state}
"""
    )

    return {
        "suggestion": response.content
    }


# ------------------------------------------------------------------
# Poem Subgraph
# ------------------------------------------------------------------

class PoemState(TypedDict, total=False):
    query: str
    draft_poem: str
    poem: str


def poem_draft(state: PoemState):
    response = model.invoke(
        f"Write a short poem about {state['query']}."
    )

    return {
        "draft_poem": response.content
    }


def poem_revise(state: PoemState):
    response = model.invoke(
        f"""
Improve this poem:

{state['draft_poem']}
"""
    )

    return {
        "poem": response.content
    }


poem_builder = StateGraph(PoemState)

poem_builder.add_node("poem_draft", poem_draft)
poem_builder.add_node("poem_revise", poem_revise)

poem_builder.add_edge(START, "poem_draft")
poem_builder.add_edge("poem_draft", "poem_revise")
poem_builder.add_edge("poem_revise", END)

poem_subgraph = poem_builder.compile()


# ------------------------------------------------------------------
# Main Graph
# ------------------------------------------------------------------

builder = StateGraph(State)

builder.add_node("joke_node", joke_node)
builder.add_node("poem_workflow", poem_subgraph)

builder.add_node("summarize", summarize)
builder.add_node("final_node", final_node)

# Run joke and poem workflow in parallel

builder.add_edge(START, "joke_node")
builder.add_edge(START, "poem_workflow")

# Fan-in

builder.add_edge("joke_node", "summarize")
builder.add_edge("poem_workflow", "summarize")

builder.add_edge("summarize", "final_node")
builder.add_edge("final_node", END)

graph = builder.compile()


# ------------------------------------------------------------------
# Stream Demo
# ------------------------------------------------------------------

stream = graph.stream_events(
    {"query": "1+1"},
    version="v3",
)

for channel, item in stream.interleave(
    "values",
    "messages",
    "subgraphs",
):
    if channel == "values":
        print("\n[state]")
        print(item)

    elif channel == "messages":
        print(
            f"\n[llm] node={item.node}"
        )

        for token in item.text:
            print(token, end="", flush=True)

    elif channel == "subgraphs":
        print(
            f"\n[subgraph] path={item.path}"
        )
````
